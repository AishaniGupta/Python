```python
#Strong Password
password=input("Enter your password: ")
i=0

upper=False
lower=False
digit=False
special_character=False

if 6<=len(password)<=16:
  
  while i!=len(password):
    if "a"<= password[i] and "z">=password[i]:
      lower=True
    elif "A"<= password[i] and "Z">=password[i]:
      upper=True
    elif "0"<=password[i] and "9">=password[i]:
      digit=True
    elif password[i]=="@" or password[i]=="#" or password[i]=="%" or password[i]=="!" or password[i]=="$" or password[i]=="&":
      special_character=True
    i+=1
else:
  if len(password)<6:
    print("Your password is too short")
  else:
    print("Your password is too long")

if upper==False:
    print("Please add a Upper Case Alphabet")
elif lower==False:
    print("Please add a Lower Case Alphabet")
elif digit==False:
    print("Please add a Numeric Character ")   
elif special_character==False:
    print("Please add a Special Character")
else:
    print("You have a perfectly Strong Password")

```

```python
#Armstrong Number
num=int(input("Enter a number: "))
number=int(input("Enter the digits of your number: "))
n=num
sum=0
i=1
while i<=number:
  d=num%10
  d=d**number
  sum=sum+d
  num=num//10
  i+=1
if n==sum:
  print("Yes its a Armstrong Number")
else:
  print("No, its not a Armstrong Number")
```

```python
# If n is odd, print Weird
# If n is even and in the inclusive range of 2 to 5, print Not Weird
# If n is even and in the inclusive range of 6 to 20, print Weird
# If n is even and greater than 20, print Not Weird

num=int(input("Enter a number: "))
while num:
  if num%2==0:
    if 2<=num<=5:
      print("Even-2 to 5-Not Weird")
    elif 6<=num<20:
      print("Even-6 to 20-Weird")
    elif num>20:
      print("Even-Greater than 20-Not Weird")
    break
  elif num%2!=0:
    print("Odd-Weird")
    break
  else:
    print("Your given number does not matches with any conditions")
```

```python
# Last number divisible
# Write a python program to find out the last digit of the number and check if it is divisible by 2 or not.
num=int(input("Enter a number: "))
while num:
  n=num%10
  if n%2==0:
    print("Yes, the last digit of the given number is divisible by 2")
    break
  else:
    print("No, the last digit of the given number is not divisible by 2")
    break
```

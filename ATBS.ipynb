```python
import random, sys

print("ROCK, PAPER SCISSORS")
wins,losses,ties=0,0,0
while True:     #MAIN LOOP OF THE GAME
  print("% win, %loss, %ties", (wins,losses,ties))
  while True:
    move=input("Enter your move: (R)ock or (P)aper or (S)cissors or (Q)uit: ")
    if move=="Q":
      sys.exit()
    if move=="R" or move=="P" or move=="S":
      break
    print("Choose between R,P,S,Q")
    
  if move=="R":
    print("ROCK VS ")
  elif move=="S":
    print("SCISSORS VS ")
  elif move=="P":
    print("PAPER VS ")
  

  randomnumber=random.randint(1,3)
  if randomnumber==1:
    Computermove= "R"
    print("ROCK")
  elif randomnumber==2:
    Computermove= "P"
    print("PAPER")
  elif randomnumber==3:
    Computermove= "S"
    print("SCISSORS")

  
  if move==Computermove:
    print("Its a tie")
    ties+=1
  elif move=="R" and Computermove=="P":
    print("You Loose")
    losses+=1
  elif move=="R" and Computermove=="S":
    print("You Win")
    wins+=1
  elif move=="P" and Computermove=="R":
    print("You Win")
    wins+=1
  elif move=="P" and Computermove=="S":
    print("You Loose")
    losses+=1
  elif move=="S" and Computermove=="P":
    print("You Win")
    wins+=1
  elif move=="S" and Computermove=="R":
    print("You Loose")
    losses+=1
```

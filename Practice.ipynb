```python
def twisted(n):
  if (isPrime(n)):
    if (isPrime(reverse(n))):
      return "Its a Twisted Prime Number"
    else:
      return "Not a Twisted Prime Number"
  else:
    return "Not a Twisted Prime Number"

    
def isPrime(n):
  i=2
  while i<n:
    if n%i==0:
      return False
    else:
      i+=1
    return True

def reverse(n):
  rev=0
  while n>0:
    lastdigit=n%10
    rev=rev*10+lastdigit
    n//=10
    

twisted(71)
```

```python
#Insertion Sort
N=int(input())
a=[0]*N
a.append(int(input()))
for i in range(1,N):
  b=a[0]
  n=int(input())
  if b<n:
    a.append(n)
  else:
    a.insert(0,n)
  for j in range(len(a)-1):
    if a[j]>a[j+1]:
      a[j],a[j+1]=a[j+1],a[j]
print(a)
```

```python
#Perfect Number- A perfect number is the one which is equal to the sum of all its factor(including 1 and excluding itself)

def perfect(num):
  sum=0
  i=1
  while i!=num:
    if num%i==0:
      sum+=i
    i+=1
  if num==sum:
    print("Yes its a Perfect Number")
  else:
    print("No its not a Perfect Number")

perfect(6)
perfect(28)
```

```python
def bmi(height,weight):
  height=(height/100)**2
  bmi=weight/height
  print(bmi,"kg/m^2")
  if bmi<=18.5:
    return "Underweight"
  elif 18.5<bmi<=25:
    return "Normal"
  elif 25<bmi<=30:
    return "Overweight"

bmi(153,43.7)
```

```python
# Write a Python function that accepts a string and calculate the number of upper case letters and
# lower case letters. Go to the editor
# Sample String : 'The quick Brow Fox'
# Expected Output :
# No. of Uppercase characters : 3
# No. of Lower case Characters : 12

def check(str=input("Enter your string: ")):
  u,l,s=0,0,0
  for i in range(len(str)):
    Z=ord(str[i])
    if 65<=Z and Z<=90:
      u+=1
    elif 97<=Z and Z<=122:
      l+=1
    else:
      s+=1
  print("No. of Uppercase characters:",u)
  print("No. of Lower case Characters:",l)
  print("No. of Special Characters:",s)

check()
```

```python
def split(l):
  a=list(l)
  b=[]
  for i in range(len(l)):
    if a[i]!=" ":   
      b.append(a[i])
  c="".join(b)
  print(c)

split("Book reading is an essential skill")
```

```python
def split_fun(in_list):
  list=[]
  start=0
  i=0
  while i<len(in_list):
    if (in_list[i]!=" ") :
        i+=1
    else:
        list.append(in_list[start:i])
        start=i+1
        i+=1
  list.append(in_list[start:])
  print(list)
split_fun("I am a girl")
```

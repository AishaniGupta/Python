```python
#Write a progrsm to check whether its a magic square or not for 15
magic_square = [[8, 3, 4],[1, 5, 9],[6, 7, 2]]

for i in range(len(magic_square)):
  sum_v,sum_h,sum_d=0,0,0
  for j in range(len(magic_square [i])):
     sum_v+=magic_square [i][j]
  # print(sum_v,end=", ")
  for j in range(len(magic_square)):
    sum_h+=magic_square[j][i]
  # print(sum_h,end=", ")
  for i in range(len(magic_square)):
    sum_d+=magic_square[i][i]
  # print(sum_d,end=", ")

if sum_v==sum_h==sum_d==15:
  print("Yes! Its a magic square")
else:
  print("No! Its not a magic square")
```

```python
p=list(input("Enter any charater containing only parenthesis: "))
l=[]
for i in range(len(p)):
  if p[i]=="(" or p[i]=="[" or p[i]=="{":
    l.append(p[i])
  else:
    if p[i]==")":
      if l[-1]=="(":
        l.pop()
    if p[i]=="]":
      if l[-1]=="[":
        l.pop()
    if p[i]=="}":
      if l[-1]=="{":
        l.pop()

if l==[]:
  print("Valid")
else:
  print("Invalid")
```

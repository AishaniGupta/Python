```python
#Pivot Element (The element at which, all the elements from the right's and left's sum is equal)
input=[1,2,8,6,4,4,3]
i=0
lsum=0
rsum=0
j=len(input)-1
while i<j:
  if lsum<rsum:
    i+=1
  elif lsum>rsum:
    j-=1
  else:
    if (i+1)==(j-1):
      print(input[i+1])
    else:
      lsum+=input[i]
      rsum+=input[j]
      i+=1
      j-=1
```

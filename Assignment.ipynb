```python
#Reverse an array
N=int(input())
l=[0]*N
for i in range(N):
  n=int(input())
  l[i]=n
print(l)
a=N//2
for j in range(a):
  l[-1-j],l[0+j]=l[0+j],l[-1-j]
  a-=1
print(l)
```

```python
#Find the mean, median and mode in an array
#ADD MODE
N=int(input())
l=[None]*N
for i in range(N):
  n=int(input())
  l[i]=n
print(l)
sum=0
for i in range(N):
  sum+=l[i]
print("Mean=",sum/N)
max,j,c=0,0,-1
d=N
while d!=0:
  for j in range(d):
    if max<l[j]:
      max=l[j]
      b=j
  l[b]=l[c]
  l[c]=max
  max=0
  d-=1
  c-=1
print(l)
a=N//2
if N%2==0:
  Median=(l[a-1]+l[a])/2
  print("Median=",Median)
else:
  print("Median=",l[a])
```

```python
# Print the elements of an array of size N, in a rotated manner with a gap of K. Eg, let us say the array is - 3, 6, 7, 5, 10. And the value of k = 3. 
# The output should be - 7,3,10,6,5. If k = 2, the output should be - 6,5,3,10,7
N=int(input())
l=[None]*N
m=[0]*N
for i in range(N):
  n=int(input())
  l[i]=n
print(l)
k,j,p,c=0,0,N,0
while p!=0:
  c+=1
  j+=1
  if c==4:
    m[k]=l[j]
    l[j],l[-1]=l[-1],l[j]
    k+=1
    p-=1
    c=0
  if j==3:
    j=0
print(m)
```

```python
# Find duplicate elements in an array
N=int(input())
l=[]
for i in range(N):
  n=int(input())
  l.append(n)
print(l)
a=0
while a<len(l):
  b=l[a]
  for j in range(a+1,len(l)):
    c=0
    if b==l[j]:
      c+=1
    if c>=1:
      print("The duplicated element in the array is=",l[j])
  a+=1
```

```python
# Merge two sorted arrays into a single sorted array
A=[3,5,6,8,9,11,15]
B=[1,3,4,5,12,17]
C=[]
i,j=0,0
N=len(A)+len(B)
while i*2<N:
  if A[i]>B[j]:
    C.append(B[j])
    j+=1
  else:
    C.append(A[i])
    i+=1
if j>i:
    C.append(A[i])
else:
    C.append(B[j])
print(C)
```

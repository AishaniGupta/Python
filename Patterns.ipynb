```python
#Hollow Triangle with while loop
rows=int(input("Enter the no. of rows you want: "))
col=(2*rows)-1
mid=col//2
i=1
while i<=rows:
  j=1
  while j<=col:
    if i==rows or i+j==mid+2 or j-i==mid :
      print("*", end=" ")
    else:
     print(" ",end=" ")
    j+=1
  i+=1
  print()
```
